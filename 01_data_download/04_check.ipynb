```python
from AlgorithmImports import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import plotly.graph_objects as go
from datetime import datetime

# ------------------------------------------------------
# 1. 初始化 QuantBook, 获取 GC 期货秒级数据
# ------------------------------------------------------
qb = QuantBook()
#symbol = qb.add_future(Futures.Metals.Gold, Resolution.DAILY).symbol
symbol = qb.add_future(Futures.Indices.NASDAQ_100_E_MINI, Resolution.SECOND).symbol

start = datetime(2024, 1, 1)
end = datetime(2024, 1, 2)
df = qb.history(symbol,
    start,
    end,
    resolution=Resolution.SECOND, 
    fill_forward=True, 
    extended_market_hours=True,
    data_mapping_mode=DataMappingMode.OPEN_INTEREST, 
    data_normalization_mode=DataNormalizationMode.BACKWARDS_RATIO, 
    contract_depth_offset=0).reset_index()

print(f"原始数据行数: {len(df)}")
print(df.head())


# 创建 Candlestick 对象
candlestick = go.Candlestick(
    x=df['time'], 
    open=df['open'], 
    high=df['high'], 
    low=df['low'], 
    close=df['close'],
    name='Candlestick'
)

# 定义图表布局 (可选，用于设置标题、轴标签等)
layout = go.Layout(
    title=f'{symbol} Candlestick Chart',
    xaxis=dict(title='Date'),
    yaxis=dict(title='Price (USD)'),
    xaxis_rangeslider_visible=False # 隐藏底部的范围选择器
)

# 创建图表对象
fig = go.Figure(data=[candlestick], layout=layout)

# 显示图表
fig.show()
```
